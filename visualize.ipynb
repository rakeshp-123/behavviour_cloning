```python
# Image visualization of the captured training image from track 1
import csv
import numpy as np
import cv2
import ntpath
    
lines = []
with open('driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        lines.append(line)
features = []
labels = []
count = 0
for line in lines:
# loop for center,left and right camera images
    for i in range(3):                         
        filename = ntpath.basename(line[i])
        current_p = 'IMG_track1/' + filename
        #print(current_p)
        im = cv2.imread(current_p)
        # convert from BGR to RGB
        im_rgb = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
        features.append(im_rgb)
    count+=1    
    if count == 8:
        break
import random
import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
# Display 8 Images

fig, ax = plt.subplots(2, 4)
ax = ax.flatten()    
for i in range(8):
    index = random.randint(0, len(features))
    ax[i].axis('off')
    ax[i].imshow(features[index])
    
        
```
